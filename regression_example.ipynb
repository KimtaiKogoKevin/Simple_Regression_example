```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
time_studies = np.array([20,50,32,65,23,43,10,5,22,35,29,5,56]).reshape(-1,1)
scores =np.array([36,49,65,30,45,60,15,35,60,2,44,20,55]).reshape(-1,1)

```

```python
model = LinearRegression()
model.fit(time_studies,scores)

print(model.predict(np.array([56]).reshape(-1,1)))
```

```python
plt.scatter(time_studies,scores)
plt.plot(np.linspace(0,70,100).reshape(-1,1),model.predict(np.linspace(0,70,100).reshape(-1,1)),'r')
plt.show()
```
